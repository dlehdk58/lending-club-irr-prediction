```python
from fredapi import Fred

fred = Fred(api_key = 'API_KEY')
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/lending_club_2020_train.csv')
```

```python
df['term'].unique()
```

```python
df['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%Y')

# Lending Club 데이터에 매칭
df['year_month'] = df['issue_d'].dt.to_period('M')
```

```python
# 3년 만기 국채 수익률 가져오기 (예시)
treasury_3yr = fred.get_series('DGS3')

# 데이터 프레임으로 변환
treasury_3df = treasury_3yr.reset_index()
treasury_3df.columns = ['date', 'treasury_rate_3yr']

# 날짜를 월 단위로 변환하여 평균 수익률 산출
treasury_3df['year_month'] = treasury_3df['date'].dt.to_period('M')
monthly_3yr = treasury_3df.groupby('year_month')['treasury_rate_3yr'].mean().reset_index()

# 5년 만기 국채 수익률 가져오기 (예시)
treasury_5yr = fred.get_series('DGS5')

# 데이터 프레임으로 변환
treasury_5df = treasury_5yr.reset_index()
treasury_5df.columns = ['date', 'treasury_rate_5yr']

# 날짜를 월 단위로 변환하여 평균 수익률 산출
treasury_5df['year_month'] = treasury_5df['date'].dt.to_period('M')
monthly_5yr = treasury_5df.groupby('year_month')['treasury_rate_5yr'].mean().reset_index()
```

```python
monthly_3yr.to_csv('./data/treasury_3yr.csv', index=False)
monthly_5yr.to_csv('./data/treasury_5yr.csv', index=False)
```
